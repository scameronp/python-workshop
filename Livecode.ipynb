```python
import seaborn as sns
```

```python
sns.countplot(data=my_dataframe, x='neighbourhood_group')
```

```python
order = my_dataframe['neighbourhood_group'].value_counts().index
sns.countplot(data=my_dataframe, x='neighbourhood_group', order=order)
```

```python
my_dataframe['room_type'].unique()
```

```python
sns.countplot(data=my_dataframe, x="neighbourhood_group", hue="room_type")
```

```python
sns.histplot(my_dataframe['price'], kde=True)
```

```python
affordable_df = my_dataframe[my_dataframe['price'] <= 500]
sns.histplot(affordable_df['price'], kde=True)
```

```python
my_dataframe.price.mean()
```

```python
affordable_df.price.mean()
```

```python
sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15, 8))

sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

```python
affordable_df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```

```python
background_image = plt.imread('https://raw.githubusercontent.com/lewagon/data-images/master/workshops/Neighbourhoods_New_York_City_Map.png')
plt.imshow(background_image, zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])

ax = plt.gca()
affordable_df.plot(
  ax=ax,
  zorder=1,
  kind='scatter',
  x='longitude',
  y='latitude',
  c='price',
  cmap='inferno',
  colorbar=True,
  alpha=0.8,
  figsize=(12,8)
)
```
